```python
import cv2
import serial
import time
import pickle
import numpy as np

serialComm = serial.Serial('COM3', 9600)
serialComm.timeout = 1

def faceFunction(picturePath):
    imagePath = picturePath
    cascPath = r'C:\Users\nuela\Desktop\Python\JupyterNoteBook\Arduino\FaceIdentification\haarcascade_frontalface_default.xml'
    faceCascade = cv2.CascadeClassifier(cascPath)

    recognizer = cv2.face.LBPHFaceRecognizer_create()
    recognizer.read("trainer.yml")

    labels = {"Person_name": 1}
    with open("labels.pickle",'rb') as f:
        original_labels = pickle.load(f)
        labels = {v:k for k,v in original_labels.items()}
        
        
    # Read the image that we want to analyze
    image = cv2.imread(imagePath)
    cv2.imshow('Original image',image) # display original image in a new windows
    cv2.waitKey(0) # waits until you press a key (any key)
    cv2.destroyAllWindows() # closes the image that was opened as a new window

    #Transform the image to Grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) # transforms the color image into grayscale to improve pattern recognition
    cv2.imshow('Gray image', gray) # display the grayscale image in a new windows
    cv2.waitKey(0) # waits until you press a key (any key)
    cv2.destroyAllWindows() # closes the gray image that was opened as a new window
    
    
    # Detect cat's faces in the image
    faces = faceCascade.detectMultiScale( # faces is a list of vectors containing the coordinates of the rectangles that encapsulate the faces found in the image
        gray,
        scaleFactor= 1.21, #1.25
        minNeighbors=5, #5
        minSize=(20, 20), #30, 30
        flags = cv2.CASCADE_SCALE_IMAGE
    )
    for (x,y,w,h) in faces:
        roi_gray = gray[y:y+h, x:x+w]
        roi_color = image[y:y+h, x:x+w]
    
        id_now , conf = recognizer.predict(roi_gray)
        if conf >= 45:
            print(id_now)
            print(labels[id_now])
            font = cv2.FONT_HERSHEY_SIMPLEX
            name = labels[id_now]
            color = (255,255,255)
            stroke = 2
            cv2.putText(image, name, (x,y), font, 1,color, stroke, cv2.LINE_AA)
   
    print('The name of person found: ', labels[id_now])
    personName = labels[id_now]

    serialComm.write(personName.encode())

    # Draw a rectangle around the faces
    for (x, y, w, h) in faces:
        cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 2)
    
    #Display the cat faces found on the image
    cv2.imshow("Faces found", image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
    
for item in range (0,7):
    if item == 1:
        faceFunction(r'C:\Users\nuela\Desktop\Python\JupyterNoteBook\Arduino\FaceIdentification\images\Nuela.jpg')
    if item == 2:
        faceFunction(r'C:\Users\nuela\Desktop\Python\JupyterNoteBook\Arduino\FaceIdentification\images\Beyonce.jpg')
    if item == 3:
        faceFunction(r'C:\Users\nuela\Desktop\Python\JupyterNoteBook\Arduino\FaceIdentification\images\BradPitt.jpg')
    if item == 4:
        faceFunction(r'C:\Users\nuela\Desktop\Python\JupyterNoteBook\Arduino\FaceIdentification\images\EmmaRoberts.jpg')
    if item == 5:
        faceFunction(r'C:\Users\nuela\Desktop\Python\JupyterNoteBook\Arduino\FaceIdentification\images\JenniferAniston.jpg')
    if item == 6:
        faceFunction(r'C:\Users\nuela\Desktop\Python\JupyterNoteBook\Arduino\FaceIdentification\images\LeonardoDicaprio.jpg')
        
```
