```python
import os
import cv2
import numpy as np
import pickle
from PIL import Image

base_dir = os.path.dirname(os.path.abspath("__file__"))
image_dir = os.path.join(base_dir, "images")
cascPath = r'C:\Users\nuela\Desktop\Python\JupyterNoteBook\Arduino\FaceIdentification\haarcascade_frontalface_default.xml'

face_cascade = cv2.CascadeClassifier(cascPath)
recognizer = cv2.face.LBPHFaceRecognizer_create()

current_id = 0
label_ids = {}
y_labels = []
x_train = []

for root, dirs, files in os.walk(image_dir):
    for file in files:
        if file.endswith("jpg") or file.endswith("png"):
            path = os.path.join(root, file)
            label = os.path.basename(path).replace(".jpg", "").replace(".png", "") #.replace(" ", "-").lower()
            print (label, path)
            
            if label not in label_ids:
                label_ids[label] = current_id
                current_id += 1
            id_now = label_ids[label]
            print(label_ids)
                           
            pil_image = Image.open(path).convert("L") #open image and convert it to grayscale.
            image_array = np.array(pil_image, "uint8")
            print(image_array)
            
            
            
            faces = face_cascade.detectMultiScale(image_array,
                scaleFactor=1.21, #1.25
                minNeighbors=4, #5
                minSize=(20, 20), #30, 30
                flags = cv2.CASCADE_SCALE_IMAGE)
            
            for (x,y,w,h) in faces:
                roi = image_array[y:y+h,x:x+w]
                x_train.append(roi)
                y_labels.append(id_now)
                
                
#print(y_labels)
#print(x_train)
                         
with open("labels.pickle",'wb') as f:
    pickle.dump(label_ids, f)
    
recognizer.train(x_train,np.array(y_labels))
recognizer.save("trainer.yml")

            
            
```
