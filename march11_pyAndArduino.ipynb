```python
import serial
import numpy as np
serialCom = serial.Serial('COM6', baudrate = 9600, timeout = 1)

allMyData = np.array([])

while True:
    arduinoData = serialCom.readline().decode('ascii')
    print(arduinoData)
    np.append(allMyData, arduinoData)
    
    if arduinoData == 'finish':
        np.savetxt('data_out', allMyData, delimeter= ',')
```
